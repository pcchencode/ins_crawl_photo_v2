```python
## ÂÖàÊäì„ÄåÂàÜÈ†Å„ÄçÈÄôÂÄãhashtagÂ∞±Â•ΩÔºåÁ∏ΩÂÖ±34ÂâáË≤ºÊñá„ÄÅÊØèÊ¨°query 10ÂâáË≤ºÊñá
# ÊúÄÂæåË≥áÊñôÈÉΩÊäìÂÆå„ÄÅÂÜçÂ°ûÈÄ≤db
tag_name = "Êàë‰πüÊòØ"
final_df = pd.DataFrame(columns=['shortcode','tag_type','_at','is_video','post_time','like','comment','content'])
has_next_page = False
while True:
    variables = {"tag_name":tag_name,"first":"200"}
    if has_next_page:
        variables['after'] = end_cursor

    variables = json.dumps(variables, separators=(',', ':'))
    post_url = "https://www.instagram.com/graphql/query/?query_hash=9b498c08113f1e09617a1703c22b2f32&variables=%s"
    print(urllib.parse.quote(variables))

    url = post_url % urllib.parse.quote(variables)
    print(url)
    Url = req.Request(url, headers=headers)
    with req.urlopen(Url) as response:
        data = response.read().decode("utf-8")
    soup = bs4.BeautifulSoup(data, "html.parser")
    js = json.loads(soup.string)

    has_next_page = js['data']['hashtag']['edge_hashtag_to_media']['page_info']['has_next_page'] # Áî®‰æÜÂà§Êñ∑ÊòØÂê¶Ë¶Å‰∏ã‰∏ÄÈ†Å
    end_cursor = js['data']['hashtag']['edge_hashtag_to_media']['page_info']['end_cursor'] # Êõ¥Êñ∞‰∏ã‰∏ÄÈ†ÅÁöÑ url
    
    len_post = len(js['data']['hashtag']['edge_hashtag_to_media']['edges'])
    print(f"Ë©≤È†ÅË≤ºÊñáÊï∏Ôºö{len_post}")
    # ÂæåÁ∫å for loop ÊêúÂ∞ãÊñáÁ´†
    for i in range(0, len_post):
        short_code = js['data']['hashtag']['edge_hashtag_to_media']['edges'][i]['node']['shortcode']
        #print(short_code)
        if short_code in old_shortcodes:
            break
        is_video = js['data']['hashtag']['edge_hashtag_to_media']['edges'][i]['node']['is_video']
        if is_video:
            is_video = 1
        else:
            is_video = 0
        post_time = js['data']['hashtag']['edge_hashtag_to_media']['edges'][i]['node']['taken_at_timestamp']
        like = js['data']['hashtag']['edge_hashtag_to_media']['edges'][i]['node']['edge_liked_by']['count']
        comment = js['data']['hashtag']['edge_hashtag_to_media']['edges'][i]['node']['edge_media_to_comment']['count']
        edges = js['data']['hashtag']['edge_hashtag_to_media']['edges'][i]['node']['edge_media_to_caption']['edges']
        if len(edges)>0:
            content = js['data']['hashtag']['edge_hashtag_to_media']['edges'][i]['node']['edge_media_to_caption']['edges'][0]['node']['text']
        else:
            content = ""
        #if 'Ë≤®' in text:
        #    print(f"Ë©≤ÁØáÁÇ∫Êé®Èä∑ÊñáÔºöhttps://www.instagram.com/p/{short_code}/")
        # ÂÖàÊúâ@‰∫∫Â∞±‰ª£Ë°®Êúâ
        if '@' in content and '@' in content:
            _at = 1
            #print("ÊúâÂá∫ÁèæÈóúÈçµ@")
        else:
            _at = 0
        # display_url = js['data']['hashtag']['edge_hashtag_to_media']['edges'][i]['node']['display_url']
        tag_type = tag_name
        
        data_list = [short_code, tag_type, _at, is_video, post_time, like, comment, content]
        final_df.loc[len(final_df.index)] = data_list
    
    if has_next_page==False:
        break
    time.sleep(random.randrange(10,30))
```
